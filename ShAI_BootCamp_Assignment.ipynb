```python
# 3. **Data Cleaning**: Handle missing data by suitable method with explain why you use it.
# Handle missing data by suitable method with explain why you use it.
df.isnull().sum()
```

```python
# get the number of unique values in each column with missing values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(col, df[col].nunique())
```

```python
# Handle missing data in 'BasePay' column by replacing missing values with mean value of that column
# Because the mean is the average, it is the most commonly used measure of central tendency.
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df.isnull().sum()
```

```python
# Handle missing data in 'OvertimePay' column by replacing missing values with mean value of that column
# Because the mean is the average, it is the most commonly used measure of central tendency.
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df.isnull().sum()
```

```python
# Handle missing data in 'OtherPay' column by replacing missing values with mean value of that column
# Because the mean is the average, it is the most commonly used measure of central tendency.
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df.isnull().sum()
```

```python
# Handle missing data in 'Benefits' column by replacing missing values with mean value of that column
# Because the mean is the average, it is the most commonly used measure of central tendency.
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
df.isnull().sum()
```

```python
# We can drop the 'Notes' and 'Status' columns because they have too many missing values
df.drop(['Notes', 'Status'], axis=1, inplace=True)
df.isnull().sum()
```

```python
# 4. **Basic Data Visualization**: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], kde=True, stat="density", kde_kws=dict(cut=3))
plt.title('Distribution of Total Pay')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar chart to visualize the average salary in each year
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='TotalPay', data=df)
plt.title('Average Salary in Each Year')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.show()

```

```python
# Create a pie chart to visualize the proportion of employees in different departments
plt.figure(figsize=(10, 6))
df['JobTitle'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.ylabel(' ')
plt.show()
```

```python
# 5. **Grouped Analysis**: Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

pd.set_option('display.max_columns', 500) # Show all columns
# Group the data by 'Year' column and calculate summary statistics for each group
df.groupby('Year').describe()

```

```python
# Group the data by 'Year' column and calculate the average 'salary' for each group
df.groupby('Year')['TotalPay'].mean()
```

```python
# Compare the average salaries across different groups with bar chart
plt.figure(figsize=(10, 6))
df.groupby('Year')['TotalPay'].mean().plot(kind='bar')
plt.title('Average Salary in Each Year')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.show()
```
